```python
import pandas as pd
import datetime
import json
```

```python
youtube_raw_df = pd.read_csv('raw_youtube_metadata/US_youtube_trending_data.csv', header=0)

# typecast column to datetime object
youtube_raw_df['trending_date'] = pd.to_datetime(youtube_raw_df['trending_date'])
youtube_raw_df['publishedAt'] = pd.to_datetime(youtube_raw_df['publishedAt'])
```

```python
youtube_raw_df.drop('thumbnail_link', axis=1, inplace=True)
youtube_raw_df.drop('comments_disabled', axis=1, inplace=True)
youtube_raw_df.drop('ratings_disabled', axis=1, inplace=True)
youtube_raw_df.drop('description', axis=1, inplace=True)

youtube_raw_df = youtube_raw_df.sort_values(by='trending_date', ascending=True)
```

```python
trending_dates = list(set(youtube_raw_df["trending_date"]))
print(type(trending_dates))
print(len(trending_dates))
print(type(trending_dates[1]))
```

```python
trending_dates = sorted(trending_dates, key=lambda x: x.date())
trending_dates_dict = {k: None for k in trending_dates}
```

```python
print(trending_dates_dict)
```

```python
# iterate over the rows of the dataframe
for index, row in youtube_raw_df.iterrows():
    if trending_dates_dict.get(row["trending_date"]) is None:
        trending_dates_dict[row["trending_date"]] = row
    else:
        if trending_dates_dict.get(row["trending_date"])['view_count'] < row["view_count"]:
            trending_dates_dict[row["trending_date"]] = row
```

```python
# Convert dictionary values to list
trending_dates_lsit = list(trending_dates_dict.values())

# Convert list to DataFrame
youtube_df = pd.DataFrame(trending_dates_lsit)
```

# Getting dictionary of Categories

```python
# Read in JSON file
with open('raw_youtube_metadata/US_category_id.json', 'r') as f:
    data = json.load(f)
```

```python
categories_dict = {}

for item in data.get('items'):
    id = item.get('id')
    category = item.get('snippet').get('title')
    categories_dict[id] = category
```

```python
print(categories_dict)
```

# Convert youtube_df categoryId column to category names

```python
for index, row in youtube_df.iterrows():
    id = row['categoryId']
    # print(id)
    category = categories_dict.get(str(id))
    # print(category)
    youtube_df.at[index, 'categoryId'] = category
```

```python
youtube_df['trending_date'] = pd.to_datetime(youtube_df['trending_date'])
```

# Combining Youtube Dataset with Master Pixstory Dataset

`Features`
1. Title of Daily Trending Youtube Video (video_id)
2. Channel Title of Daily Trending Youtube Video (channelId)
3. Category Type (categoryId)
4. View Count
5. Likes Count
6. Dislikes Count
7. Comments Count
8. Likes-to-Dislikes Ratio
9. Views-to_Likes Ratio

```python
df_master_dataset = pd.read_csv("../Master_Dataset_Raw.csv")
# convert datetime
df_master_dataset['Account Created Date'] = pd.to_datetime(df_master_dataset['Account Created Date'])
df_master_dataset['Account Created Date'] = pd.to_datetime(df_master_dataset['Account Created Date'].dt.date)

youtube_df['trending_date'] = pd.to_datetime(youtube_df['trending_date'])
youtube_df['trending_date'] = pd.to_datetime(youtube_df['trending_date'].dt.date)

youtube_df['trending_date'] = youtube_df['trending_date'].astype(str)
df_master_dataset['Account Created Date'] = df_master_dataset['Account Created Date'].astype(str)
```

```python
for index, row in df_master_dataset.iterrows():

    youtube_vid = youtube_df[youtube_df['trending_date'] == row['Account Created Date']]
    youtube_vid = youtube_vid.values.tolist()

    try:
        # print(index, "**********\n", title[0][1])
        df_master_dataset.loc[index, 'YouTube Trending Video Title'] = youtube_vid[0][1]
        df_master_dataset.loc[index, 'YouTube Trending Video Channel'] = youtube_vid[0][4]
        df_master_dataset.loc[index, 'YouTube Trending Video Category'] = youtube_vid[0][5]
        df_master_dataset.loc[index, 'YouTube Trending Video Views'] = youtube_vid[0][8]
        df_master_dataset.loc[index, 'YouTube Trending Video Likes'] = youtube_vid[0][9]
    finally:
        continue
```

```python
df_master_dataset.to_csv('Master_Dataset_Raw_Youtube.csv', index=False)
# youtube_df.to_csv('youtube.csv', index=False)
```
